```python
# 1. Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LinearRegression

# 2. Read and preprocess the dataset
df = pd.read_csv("student_lifestyle_dataset..csv")
df.dropna(inplace=True)

# 2.1 Print the original grade distribution to check if there are multiple grades
print("Original grade distribution:")
print(df['Grades'].value_counts())  # Check the distribution of the original grades
print("\nOriginal grade descriptive statistics:")
print(df['Grades'].describe()) # Check descriptive statistics

# 2.2 Convert 'Grades' to categorical performance labels
def grade_to_category(grade):
    # Adjusting thresholds based on the data
    if grade < 6:  # Fail if less than 6
        return "Fail"
    elif grade < 7:  # Pass if less than 7
        return "Pass"
    elif grade < 8:  # Good if less than 8
        return "Good"
    elif grade < 9:  # Very Good if less than 9
        return "Very Good"
    else:  # Excellent for 9 and above
        return "Excellent"

# Apply the categorization to the 'Grades' column
y = df["Grades"].apply(grade_to_category)

# 2.3 Print the new distribution of grades after categorization
print("Grade distribution after categorization:")
print(y.value_counts())  # Check if categories contain multiple classes

# 2.4 If there is only one class in y, raise an error
if len(y.unique()) <= 1:
    print("Error: All data points belong to a single class")
    raise ValueError("The dataset contains only one class, cannot proceed with classification.")

# Drop unnecessary columns
X = df.drop(["Grades", "Student_ID"], axis=1)

# One-hot encode categorical features
X = pd.get_dummies(X)

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Train-test split (with stratification to preserve class balance)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# 3.1 Print class distribution in train and test sets to verify stratification
print("\nTraining set class distribution:")
print(y_train.value_counts())
print("\nTest set class distribution:")
print(y_test.value_counts())

# 4. Save split files
pd.DataFrame(X_train, columns=X.columns).to_csv("X_train.csv", index=False)
pd.DataFrame(X_test, columns=X.columns).to_csv("X_test.csv", index=False)
pd.DataFrame(y_train).to_csv("Y_train.csv", index=False)
pd.DataFrame(y_test).to_csv("Y_test.csv", index=False)

# 5. Train models and evaluate
def evaluate_and_save(model, name):
    # Check if the model is LinearRegression, which is not suitable for classification
    if isinstance(model, LinearRegression):
        print(f"\nðŸš« Skipping LinearRegression as it's not a classification model.")
        return None # Return None or some indicator that evaluation was skipped

    model.fit(X_train, y_train)
    preds = model.predict(X_test)

    # Save predictions
    pd.DataFrame(preds, columns=["Prediction"]).to_csv(f"predictions_{name}_model.csv", index=False)

    # Accuracy
    acc = accuracy_score(y_test, preds)
    print(f"\nâœ… {name} Accuracy: {acc:.2%}")

    # Confusion matrix - ensure labels are passed to confusion_matrix
    cm = confusion_matrix(y_test, preds, labels=y.unique()) # Use all unique labels from original y
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=y.unique(), yticklabels=y.unique())
    plt.title(f"{name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    # Classification report
    report = classification_report(y_test, preds, output_dict=True)
    df_report = pd.DataFrame(report).transpose()

    # Exclude rows that are not per-class metrics (e.g., 'accuracy', 'macro avg', 'weighted avg')
    report_metrics = ['precision', 'recall', 'f1-score']
    df_report_filtered = df_report.loc[y.unique(), report_metrics]

    if not df_report_filtered.empty:
        df_report_filtered.plot(kind="bar", figsize=(8, 4))
        plt.title(f"{name} - Classification Report")
        plt.ylabel("Score")
        plt.ylim(0, 1)
        plt.grid(axis="y")
        plt.show()
    else:
        print(f"Could not generate classification report for {name}. Check if predictions have classes present in y_test.")

    return acc

# Define classification models
models = {
    "KNN": KNeighborsClassifier(),
    "ANN": MLPClassifier(max_iter=1000),
    "SVM": SVC(),
    "RandomForest": RandomForestClassifier(),
    "DecisionTree": DecisionTreeClassifier(),
    "NaiveBayes": GaussianNB(),
    "LinearRegression": LinearRegression()  # Added LinearRegression here
}

accuracies = {}

# Train and evaluate each model
for name, model in models.items():
    print(f"\nðŸš€ Training {name}...")
    acc = evaluate_and_save(model, name)
    if acc is not None:
        accuracies[name] = acc

# Final comparison
print("\nðŸ“Š Final Accuracy Comparison:")
for model_name, acc in sorted(accuracies.items(), key=lambda x: x[1], reverse=True):
    print(f"{model_name}: {acc:.2%}")
```
